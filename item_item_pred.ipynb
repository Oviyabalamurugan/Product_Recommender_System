Colaborative Filtering (Item-Item Recommendation):

> Aim to fill in the missing entries of a user-item association matrix.





```python
new_df=rev_pos.groupby("reviewerID").filter(lambda x:x['rating'].count() >= 30) 
display(new_df.shape)
display(new_df)
```

Model Based  CF:

```python

ratings_matrix = new_df.pivot_table(values='rating', index='reviewerID', columns='asin', fill_value=0)
display(ratings_matrix)
print('shape of the ratings matrix = ', ratings_matrix.shape)
```

```python
from sklearn.model_selection import train_test_split
train = ratings_matrix.to_numpy()
sparsity = float(len(train.nonzero()[0]))
sparsity /= (train.shape[0] * train.shape[1])
sparsity *= 100
print ('Sparsity: {:5.2f}%'.format(sparsity))
```

```python
ratings_matrix = ratings_matrix.T
ratings_matrix
```

```python
X = ratings_matrix
```

```python
from sklearn.decomposition import TruncatedSVD

SVD = TruncatedSVD(n_components=10)
decomposed_matrix = SVD.fit_transform(X)
display(decomposed_matrix.shape)
display(decomposed_matrix)
```

```python
correlation_matrix = np.corrcoef(decomposed_matrix)

display(correlation_matrix.shape)
```

```python
cor = pd.DataFrame(correlation_matrix, index = X.index,columns=X.index)
display(cor)
```
